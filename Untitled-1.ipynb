````python
# RMSF Calculation Script (C1' Version)

# Purpose:

# Calculate RMSF of each nucleotide using the C1' atom.

#

# Workflow:

# DCD + PSF

# ↓

# Alignment using G4 Core

# ↓

# RMSF calculation using C1'

# ↓

# Save as NPZ

#

# Author: MYT Lab

# ==================================================

# ==================================================

# Part 1. Import Libraries

# ==================================================

import mdtraj as md
import numpy as np

# ==================================================

# Part 2. Define G4 Core Region

# ==================================================

#

# Purpose:

# Define residues used for structural alignment.

#

# Notes:

# RMSF is calculated after alignment.

# Using G-tetrad residues helps remove

# overall translation and rotation.

#

# ==================================================

CORE_RESIDUES = """
resid 2 or resid 3 or resid 4
or resid 8 or resid 9 or resid 10
or resid 14 or resid 15 or resid 16
or resid 20 or resid 21 or resid 22
"""

# ==================================================

# Part 3. System Setup

# ==================================================

#

# Purpose:

# Define topology and trajectory files.

#

# How to modify:

# Change topo path

# Change trajectory paths

#

# ==================================================

system_setup = {

```
"blank": {

    "topo":
    "/ceph/sharedfs/work/MYTLab/asher/center_ion_project/blank/na/step3_input.psf",

    "traj_list": [

        "/ceph/sharedfs/work/MYTLab/asher/center_ion_project/blank/na/rep1/G4_ion_md_na.dcd",

        "/ceph/sharedfs/work/MYTLab/asher/center_ion_project/blank/na/rep2/G4_ion_md_na.dcd",

        "/ceph/sharedfs/work/MYTLab/asher/center_ion_project/blank/na/rep3/G4_ion_md_na.dcd"

    ],

    "label": "Blank",
    "color": "green"

}
```

}

# ==================================================

# Part 4. Output File

# ==================================================

output_file = "rmsf_data_na_all_blank_c1.npz"

# ==================================================

# Part 5. Initialize Containers

# ==================================================

results = {}
system_meta = {}
labels_cache = None

print("\n====================================")
print(" RMSF Analysis Started")
print(" Alignment : G4 Core")
print(" RMSF Atom : C1'")
print("====================================\n")

# ==================================================

# Part 6. Loop Through Systems

# ==================================================

for sys_key, sys_info in system_setup.items():

```
print(f"Processing system: {sys_key}")

dcd_files = sys_info.get("traj_list")

if not dcd_files:
    continue

rmsf_replicates = []

try:

    # ==========================================
    # Part 6-1. Load Reference Frame
    # ==========================================

    ref_traj = md.load(
        dcd_files[0],
        top=sys_info["topo"],
        stride=100
    )

    ref_frame = ref_traj[0]

    print("Reference frame loaded")


    # ==========================================
    # Part 6-2. Alignment Atom Selection
    # ==========================================
    #
    # Alignment atoms:
    #   P
    #   C1'
    #   C4'
    #
    # Only G4 core residues are used.
    #
    # ==========================================

    align_selection = ref_frame.topology.select(

        f"({CORE_RESIDUES}) and "
        "(name P or name \"C1'\" or name \"C4'\")"

    )

    if len(align_selection) == 0:

        print(
            "[Warning] Core selection empty."
            " Using all nucleic backbone atoms."
        )

        align_selection = ref_frame.topology.select(
            "name P or name \"C1'\" or name \"C4'\""
        )

    print(
        f"Alignment atoms selected: "
        f"{len(align_selection)}"
    )


    # ==========================================
    # Part 6-3. RMSF Atom Selection
    # ==========================================
    #
    # RMSF calculated using:
    #   C1'
    #
    # One RMSF value per residue.
    #
    # ==========================================

    calc_selection = ref_frame.topology.select(
        "name \"C1'\""
    )

    if len(calc_selection) == 0:

        print(
            "[ERROR] No C1' atoms found."
        )

        print(
            "Check whether topology uses:"
        )

        print(
            "C1' or C1* naming convention."
        )

    print(
        f"RMSF atoms selected: "
        f"{len(calc_selection)}"
    )


    # ==========================================
    # Part 6-4. Process Replicates
    # ==========================================

    for replica_id, dcd in enumerate(
        dcd_files,
        start=1
    ):

        print(
            f"Loading Replica {replica_id}"
        )

        traj = md.load(
            dcd,
            top=sys_info["topo"]
        )

        # Alignment

        traj.superpose(

            ref_frame,

            atom_indices=align_selection

        )

        # RMSF

        rmsf = md.rmsf(

            traj,
            traj,

            atom_indices=calc_selection

        )

        rmsf_replicates.append(rmsf)

        print(
            f"Replica {replica_id} finished"
        )


    # ==========================================
    # Part 6-5. Save RMSF Data
    # ==========================================

    results[sys_key] = np.array(
        rmsf_replicates
    )

    system_meta[sys_key] = {

        "label":
        sys_info["label"],

        "color":
        sys_info["color"]

    }


    # ==========================================
    # Part 6-6. Create Residue Labels
    # ==========================================

    if labels_cache is None:

        selected_atoms = [

            traj.topology.atom(idx)

            for idx in calc_selection

        ]

        labels_cache = np.array(

            [

                f"{atom.residue.name}"
                f"{atom.residue.index+1}"

                for atom in selected_atoms

            ]

        )

    print(
        f"{sys_key} completed.\n"
    )

except Exception as e:

    print(
        f"[ERROR] {sys_key}: {e}"
    )

    import traceback
    traceback.print_exc()
```

# ==================================================

# Part 7. Save Results

# ==================================================

if results:

```
np.savez(

    output_file,

    results=results,
    labels=labels_cache,
    system_meta=system_meta

)

print("\n====================================")
print("RMSF calculation completed")
print(f"Output : {output_file}")
print("====================================\n")
```

else:

```
print(
    "No RMSF data generated."
)
```
````
